#### EDA
***DATA ANALYSIS***

**univariate analysis**

```python
#price distribution
data['price']
```

```python
#identying outlier in price
df = data[data['price']<1500]

sns.boxplot(data=df, x='price')
plt.show()
```

```python
 #price distribution

plt.figure(figsize=(4,3))
sns.histplot(data=df, x='price', bins=100)
plt.title("price distribution")
plt.ylabel("frequency")
plt.show()
```

```python
sns.__version__
```

```python
x=np.random.rand(50)
y=np.random.rand(50)
plt.scatter(x,y,color="red")
plt.title('scatter plot')
plt.show()
```

```python
df.dtypes
```

```python
plt.figure(figsize=(8,5))
sns.histplot(data=df, x='availability_365',)
plt.title("availability_365 distribution")
plt.ylabel("frequency")
plt.show()
```

```python
df.groupby(by='neighbourhood_group')['price'].mean()
```

**Feature Engineering**

```python
#price per bed
df['price per bed']=df['price']/df['beds']
df.head()
```

```python
df.groupby(by='neighbourhood_group')['price per bed'].mean()
```

**Bi variable analysis**

one variable depency in another variable

```python
df.columns
```

```python
#price dependency by neighborhood 
sns.barplot(data=df, x = 'neighbourhood_group', y = 'price', hue='room_type')
plt.show()
```

```python
sns.barplot(data=df, x = 'neighbourhood_group', y = 'price per bed')
plt.show()
```

```python
#number of reviews and price
plt.figure(figsize=(5,4))
plt.title("review by locality")
sns.scatterplot(data=df, x='number_of_reviews', y='price', hue='neighbourhood_group')
plt.show()
```

```python
sns.pairplot(data=df,vars=['price','minimum_nights','number_of_reviews','availability_365'], hue='room_type')
plt.show()
```

```python
#geographical distributionsof Airbnb 
plt.figure(figsize=(6,5))
sns.scatterplot(data=df, x='latitude', y='longitude', hue='room_type')
plt.title('geographical distributionsof Airbnb')
plt.show()
```

```python
df.dtypes
```

```python
#heat map - correlation of one variable with other with numerical columns
corr = df[['latitude','latitude','price','minimum_nights','number_of_reviews','reviews_per_month','availability_365']].corr()
corr
```

```python
corr = df[['latitude','price','minimum_nights','number_of_reviews','reviews_per_month','availability_365']].corr()
corr
sns.heatmap(data=corr, annot=True)
plt.figure(figsize=(5,5))
plt.show()
```
